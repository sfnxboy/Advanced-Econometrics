(4) Generate a 10 × 5 and a 5 × 10 array of matrices by randomly drawing observations from the standard normal distribution. Denote them A and B, respectively. Calculate C = A × B and replace diagonal elements of C to ones.

```python
A = np.random.normal(0, 1, size=(10, 5))
B = np.random.normal(0, 1, size=(5, 10))
C = A @ B

rows, columns = C.shape

for i in range(rows):
    C[i, i] = 1
    
C
```

(5) Generate a 20 × 15 array D using A in (4), and a 15 × 20 array E using B in (4).
Calculate F = D × E. Replace nonpositive (≤ 0) elements of F to 0.5

```python
D = np.hstack((np.hstack((np.vstack((A, A)), np.vstack((A, A)))), np.vstack((A, A))))
print(D.shape)
```

```python
E = np.vstack((np.hstack((np.vstack((B, B)), np.vstack((B, B)))), np.hstack((B, B))))
print(E.shape)
```

```python
F = D @ E

for i in range(len(F)):
    for j in range(len(F[i])):
        if F[i, j] <= 0:
            F[i, j] = 0.5

F[:5, :5]
```

(9) Generate the provided matrices and use `hstack`, `vstack` and `tile` to generate m. Find the shape and
dimension of m. Extract z' from m.

```python
# generating matrices
x = np.matrix([[16, 10], [13, 11]])
y = np.matrix([5])
z = np.matrix([[16, 10], [13, 11],[12, 33]])

m_top_half = np.hstack((x, np.vstack((np.hstack((np.hstack((y, y)), y)), np.hstack((np.hstack((y, y)), y))))))
print(m_top_half)

m_lower_half = np.hstack((z, np.vstack((np.transpose(z), np.hstack((y, np.hstack((y, y))))))))
print(m_lower_half)
```

```python
# all together
m = np.matrix(np.vstack((m_top_half, m_lower_half)))
print(m.shape)
print(m)
```

```python
# Extract z' from m.
import pandas as pd
m_df = pd.DataFrame(m)

print(np.transpose(z))
z_transpose_from_m = m_df.iloc[2:4, 2:]
z_transpose_from_m
```
