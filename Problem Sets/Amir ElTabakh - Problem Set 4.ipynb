In univariate time-series modeling, one class of models is known as autoregressive (AR)
models. For example, an AR(2) process can be written as

$y_t = \alpha_1 y_{t-1} + \alpha_2 y_{t-2} + u_t$

for $t = 1, 2, . . . , T$. The subscript $t$ denotes time periods and $T$ is the last period. Now, suppose $\alpha_1 = 0.8$, $\alpha_2 = 0.15$ and $u_t$â€™s are independently and identically distributed standard normal random variables. Consider the code snippet below that simulates the AR(2) for 200 periods and keeps the last 100 periods.

```python
import numpy as np
np.random.seed(37)
import matplotlib.pyplot as plt
plt.style.use('ggplot')

T = 200
alpha1 = 0.8
alpha2 = 0.15
u = np.random.randn(T)
y = np.zeros(T)
for t in range(2, T):
    y[t] = alpha1*y[t-1] + alpha2*y[t-2] + u[t]

y = y[100:] # keep the last 100 periods
```

Why do you think the for loop started from index 2? 

Answer: You need elements at index 0 and 1 in the function.

Write a while loop which generates 100 observations from the AR(2) process defined above, and then generates a line plot of the simulated AR(2) process as in Figure 1.

```python
# importing dependecies
import seaborn as sns
sns.set(rc={"figure.figsize":(10, 6)}) #width=10, #height=6

# x-axis
periods = np.arange(2, 102)

# generating plot
sns.lineplot(x = periods, y = y, color = "blue")
```

3. Consider the following code snippet. The object letters is a set containing lower case letters. Write a control structure that adds vowels to the set vowel, and consonants to the set consonant. Note that the vowels are {a,e,i,o,u}, and the rest of the letters are consonants.

```python
from string import ascii_lowercase
letters = set(ascii_lowercase)
vowel = set()
consonant = set()
all_vowels = ['a', 'e', 'i', 'o', 'u']


for letter in letters:
    if letter in all_vowels:
        vowel.add(letter)
    else:
        consonant.add(letter)

print(vowel)
print(consonant)
```

```python
x = np.arange(0, 10, 0.1)
sinx = np.sin(x)
cosx = np.cos(x)

# importing dependecies
import matplotlib.pyplot as plt
import matplotlib.path as mpath

# generating ellipse
circle = mpath.Path.unit_circle()
verts = np.copy(circle.vertices)
verts[:, 0] *= 1.3
ellipt_marker = mpath.Path(verts, circle.codes)

plt.figure(figsize=(10, 6))
plt.plot(x, sinx, label='sin(x)', color = 'green', marker = ellipt_marker,
         ms = 7, linestyle = 'None', markeredgecolor = "black")
plt.plot(x, cosx, label='cos(x)', color = 'red', marker = "D",
         ms = 5, linestyle = 'None', markeredgecolor = "black")
plt.xlabel('x')
plt.ylabel('sin(x)/cos(x)')
plt.title("Sine and Cosine Functions")
plt.legend(loc = 3)
```
